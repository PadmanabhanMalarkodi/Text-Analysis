```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
def preprocess(text):
    doc = nlp(text.lower())
    tokens = [token.lemma_ for token in doc if not token.is_punct and not token.is_stop]
    return tokens
```

```python
df['tokens'] = df['Article'].apply(preprocess)
```

```python
df
```

```python
from gensim.models import LdaModel
from gensim.corpora import Dictionary
```

```python
dictionary = Dictionary(df['tokens'])
```

```python
dictionary
```

```python
print(dictionary.token2id)
```

```python
# Create a Bag of Words (BoW) representation of the documents
corpus = [dictionary.doc2bow(tokens) for tokens in df['tokens']]
```

```python
corpus
```

```python
len(df['tokens'][0])
```

```python
df['tokens'][0]
```

```python
# Build the LDA model
num_topics = 5  # Specify the number of topics
lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=15)
```

```python
# Print the topics and their top words
topics = lda_model.print_topics(num_words=5)
for topic in topics:
    print(topic)
```

# Named Entity Recognition

```python
df['NER'] = df['Article'].apply(lambda text: [(ent.text, ent.label_) for ent in nlp(text).ents])
```

```python
df[['Article', 'NER']]
```

# Sentiment Analysis

```python
from textblob import TextBlob
```

```python
text = "The movie was good"
```

```python
TextBlob(text).sentiment.polarity
```

```python
df['Sentiment'] = df['Article'].apply(lambda text: TextBlob(text).sentiment.polarity)
```

```python
df
```

```python
df['Sentiment_Category'] = df['Sentiment'].apply(
    lambda score: 'Positive' if score > 0 else ('Negative' if score < 0 else 'Neutral')
)
```

```python
df[['Article', 'Sentiment', 'Sentiment_Category']]
```
